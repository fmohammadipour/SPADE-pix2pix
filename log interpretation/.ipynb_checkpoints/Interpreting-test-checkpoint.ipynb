```python
import pandas as pd

# Load the CSV files
file_path_without = 'wout-test_metrics_log.csv'
file_path_withsq = 'with-test_metrics_log.csv'

metrics_without = pd.read_csv(file_path_without)
metrics_withsq = pd.read_csv(file_path_withsq)
```

```python
# Display the first few rows of each dataframe to understand their structure
metrics_without.head() 
```

```python
# Display the first few rows of each dataframe to understand their structure
metrics_withsq.head()
```

```python
import matplotlib.pyplot as plt

# Extract relevant columns for plotting
id_without = metrics_without['image_id']
ssim_without = metrics_without['ssim']
psnr_without = metrics_without['psnr']
fid_without = metrics_without['fid']

id_withsq = metrics_withsq['image_id']
ssim_withsq = metrics_withsq['ssim']
psnr_withsq = metrics_withsq['psnr']
fid_withsq = metrics_withsq['fid']

# Plot SSIM, PSNR, and FID for both datasets
plt.figure(figsize=(18, 6))

# SSIM
plt.subplot(1, 3, 1)
plt.plot(id_without, ssim_without, label='Original Images', marker='o')
plt.plot(id_withsq, ssim_withsq, label='Modified Images', marker='o')
plt.xlabel('iamges')
plt.ylabel('SSIM')
plt.title('SSIM over Epochs')
plt.legend()

# PSNR
plt.subplot(1, 3, 2)
plt.plot(id_without, psnr_without, label='Original Images', marker='o')
plt.plot(id_withsq, psnr_withsq, label='Modified Images', marker='o')
plt.xlabel('images')
plt.ylabel('PSNR')
plt.title('PSNR over Epochs')
plt.legend()

# FID
plt.subplot(1, 3, 3)
plt.plot(id_without, fid_without, label='Original Images', marker='o')
plt.plot(id_withsq, fid_withsq, label='Modified Images', marker='o')
plt.xlabel('images')
plt.ylabel('FID')
plt.title('FID over Epochs')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#import pandas as pd

# Load the CSV files
file_path_without = 'wout-train-TF_metrics.csv'
file_path_withsq = 'with-train-TF_metrics.csv'

TF_metrics_without = pd.read_csv(file_path_without)
TF_metrics_withsq = pd.read_csv(file_path_withsq)

# Rename the unnamed column (assuming it's named 'Unnamed: 0')
TF_metrics_without.rename(columns={'Unnamed: 0': 'TF_epoch'}, inplace=True)
TF_metrics_withsq.rename(columns={'Unnamed: 0': 'TF_epoch'}, inplace=True)
```

```python
# Display the first few rows of each dataframe to understand their structure
TF_metrics_without.head()
```

```python
# Display the first few rows of each dataframe to understand their structure
TF_metrics_withsq.head()
```

```python
import matplotlib.pyplot as plt

# Extract relevant columns for plotting
epochs_without = TF_metrics_without['TF_epoch']
GAN_without = TF_metrics_without['GAN']
GAN_Feat_without = TF_metrics_without['GAN_Feat']
VGG_without = TF_metrics_without['VGG']
D_Fake_without = TF_metrics_without['D_Fake']
D_real_without = TF_metrics_without['D_real']

epochs_withsq = TF_metrics_withsq['TF_epoch']
GAN_withsq = TF_metrics_withsq['GAN']
GAN_Feat_withsq = TF_metrics_withsq['GAN_Feat']
VGG_withsq = TF_metrics_withsq['VGG']
D_Fake_withsq = TF_metrics_withsq['D_Fake']
D_real_withsq = TF_metrics_withsq['D_real']

# Plot SSIM, PSNR, and FID for both datasets
plt.figure(figsize=(12, 8))

# GAN
plt.subplot(2, 3, 1)
plt.plot(epochs_without, GAN_without, label='Original Images', marker='o')
plt.plot(epochs_withsq, GAN_withsq, label='Modified Images', marker='o')
plt.xlabel('Epoch')
plt.ylabel('GAN')
plt.title('GAN over Epochs')
plt.legend()

# GAN_Feat
plt.subplot(2, 3, 2)
plt.plot(epochs_without, GAN_Feat_without, label='Original Images', marker='o')
plt.plot(epochs_withsq, GAN_Feat_withsq, label='Modified Images', marker='o')
plt.xlabel('Epoch')
plt.ylabel('GAN_Feat')
plt.title('GAN_Feat over Epochs')
plt.legend()

# VGG
plt.subplot(2, 3, 3)
plt.plot(epochs_without, VGG_without, label='Original Images', marker='o')
plt.plot(epochs_withsq, VGG_withsq, label='Modified Images', marker='o')
plt.xlabel('Epoch')
plt.ylabel('VGG')
plt.title('VGG over Epochs')
plt.legend()

# D_Fake
plt.subplot(2, 3, 4)
plt.plot(epochs_without, D_Fake_without, label='Original Images', marker='o')
plt.plot(epochs_withsq, D_Fake_withsq, label='Modified Images', marker='o')
plt.xlabel('Epoch')
plt.ylabel('D_Fake')
plt.title('D_Fake over Epochs')
plt.legend()

# D_real
plt.subplot(2, 3, 5)
plt.plot(epochs_without, D_real_without, label='Original Images', marker='o')
plt.plot(epochs_withsq, D_real_withsq, label='Modified Images', marker='o')
plt.xlabel('Epoch')
plt.ylabel('D_real')
plt.title('D_real over Epochs')
plt.legend()

# Hide the last subplot (6th) since we don't need it
plt.subplot(2, 3, 6)
plt.axis('off')
plt.tight_layout()
plt.show()
```
